```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch.nn.utils.prune as prune

model_id = 'MLP-KTLim/llama3-Bllossom'

# 토크나이저 로드
tokenizer = AutoTokenizer.from_pretrained(model_id)

# 모델을 CPU에서 로드
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype=torch.bfloat16,
    device_map=None  # device_map 비활성화하여 CPU에서 로드
)
model.to("cpu")
model.eval()

# 프루닝 적용 함수
def apply_pruning(model, amount=0.3):
    """
    모델에 pruning 적용 (amount=0.5는 50% 프루닝을 의미).
    """
    for name, module in model.named_modules():
        # Linear 레이어에만 프루닝 적용
        if isinstance(module, torch.nn.Linear):
            # L1 비율로 비구조적(unstructured) 프루닝 적용
            prune.l1_unstructured(module, name='weight', amount=amount)
            prune.remove(module, 'weight')  # 프루닝 적용 후 가중치에 바로 반영

# 모델에 프루닝 적용 (50% 비율로)
apply_pruning(model, amount=0.3)

# 경량화된 모델 저장
save_path = "/content/drive/MyDrive/conference/task2 - Llama-3-Bllossom/REDUCE_MODEL/pruned_llama3_blossom_model.pth"
torch.save(model.state_dict(), save_path)

print(f"Pruned model saved at {save_path}")
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch.nn.utils.prune as prune

model_id = 'MLP-KTLim/llama3-Bllossom'

# 토크나이저 로드
tokenizer = AutoTokenizer.from_pretrained(model_id)

# 모델을 CPU에서 로드
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype=torch.bfloat16,
    device_map=None  # device_map 비활성화하여 CPU에서 로드
)
model.to("cpu")
model.eval()
def apply_pruning(model, amount=0., skip_layers=['embed_tokens', 'lm_head']):
    """
    모델에 pruning 적용 (amount=0.5는 50% pruning을 의미).
    skip_layers 리스트에 있는 레이어는 pruning에서 제외.
    """
    for name, module in model.named_modules():
        # 특정 레이어는 pruning 제외
        if any(skip_layer in name for skip_layer in skip_layers):
            continue  # 이 레이어는 pruning에서 제외

        # Linear 레이어에만 pruning 적용
        if isinstance(module, torch.nn.Linear):
            # L1 비율로 비구조적(unstructured) pruning 적용
            prune.l1_unstructured(module, name='weight', amount=amount)
            prune.remove(module, 'weight')  # pruning 적용 후 가중치에 반영

# 사용 예시
apply_pruning(model, amount=0.25, skip_layers=['embed_tokens', 'lm_head'])

torch.save(model, "/content/drive/MyDrive/conference/task2 - Llama-3-Bllossom/REDUCE_MODEL/pruned_llama3_blossom_model2.pth")
```

```python
model_id = 'MLP-KTLim/llama3-Bllossom'

# quantization 옵션 설정
bnb_config = BitsAndBytesConfig(
	#입력값을 4bit로 변환
    load_in_4bit=True,
    #모델을 4bit로 양자화
    bnb_4bit_quant_type="nf4",
    #4bit계산에 사용될 데이터유형, 4비트 부동소수점(bfloat16), 4비트정수(uint8)
    bnb_4bit_compute_dtype=torch.bfloat16
)

# 7B모델을 4비트로 양자화 하여 가져오니 1/8사이즈로 가져오는 것
model = AutoModelForCausalLM.from_pretrained(model_id, quantization_config=bnb_config)

#토크나이저는 원래처럼 가져오면 됩니다.
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
from llama_cpp import Llama

llm = Llama.from_pretrained(
	repo_id="Kimsangwook/llama3_bllossom_gguf_q5",
	filename="llama3_bllossom_gguf_q5.gguf",
)
```

```python
# 비유적 답변을 생성하기 위한 프롬프트와 지시사항 설정
PROMPT = '''당신은 창의적이고 유용한 AI 어시스턴트입니다. 사용자의 질의에 대해 비유와 은유를 통해 깊이 있는 답변을 해야 합니다, output은 전부 다 한국말로 해.'''
instruction = "분석적 사고, 최적화, 균형, 문제 해결, 효율성에 대한 비유적 표현을 생성해줘"

# 메시지 설정
messages = [
    {"role": "system", "content": PROMPT},
    {"role": "user", "content": instruction}
]

# 질문 수행 및 답변 추출
response = llm.create_chat_completion(messages=messages)

# 답변 텍스트만 출력
answer = response['choices'][0]['message']['content']
print(answer)
```
